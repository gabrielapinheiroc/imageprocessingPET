We will use the Histogram Equalization in this image. This method is usually used in Image Processing

Using:
* python
* OpenCV
* numpy

img_pet.jpeg

```python
import cv2
import numpy as np
from google.colab.patches import cv2_imshow
```

Loading the image:

```python
img_original = cv2.imread('/content/img_pet.jpeg',0)
```

Using the `equalizeHist` function we can apply the histogram equalization to our image.

```python
img_equ = cv2.equalizeHist(img_original)
```

Now, we will merge the two images and compare the equalization result.

```python
res = np.hstack((img_original,img_equ))
```

```python
cv2_imshow(res)
```

Therefore, we can conclude that the second image has more contrast after applying the method.
